<h3>Ćwiczenie 2. Brakujące wartości</h3>

<h5>Pobranie zbioru danych dotyczących rejsu Titanic ze strony: https://www.openml.org/d/40945</h5>

```python
import pandas as pd
import numpy as np
```

<h5>Dokonanie wstępnej analizy pierwszych 20 wierszy</h5>

```python
df = pd.read_csv('Titanic.csv')
df.head(20)
```

Dane składają się z 14 kolumn, między innymi takie jak: age, sex, name, survived. Wśród wartości są dane ilościowe oraz jakościowe. Widoczne braki w danych w postaci znaku '?'

<h5>Uzycie funkcji: pd.isnull().sum() oraz pd.isnull().mean w celu sprawdzenia brakujących danych.</h5>

Zbiór danych nalezy zmodyfikować poprzez zmianę znaku '?' reprezentującego braki w danych na wartość na.nan, aby zostały one wykryte przez funkcję .isnull()

```python
df = df.iloc[:,:].replace('?', np.nan)
df.head(10)
```

Obliczneie ilości nan dla poszczególnych kolumn

```python
nan_sum = df.isnull().sum()
nan_sum
```

Oblicznie procentowej ilości nan w konkretnych kolumnach dla wszystkich danych w kolumnie

```python
nan_sum/df.shape[0]*100 #df.shape[0] - liczba wierszy
```

Policzenie średniej ilości nan-ów w kazdej kolumnie - wynikiem będzie stosunek ilości nan-ów do wierszy w kolumnie z występującymi danymi

```python
df.isnull().mean()
```

Obliczenie jaki procent w dataframe stanowią braki danych

```python
nan_sum.sum()/df.size*100
```

<h5>Określenie typu braków danych: całkowicie przypadkowe (MCAR), przypadkowe (MAR), nieprzypadkowe (MNAR).</h5>

```python
df['bodyNull'] = np.where(df['body'].isnull(), 1, 0)
df
```

```python
df.groupby(['survived'])['bodyNull'].mean()
```

W szczegółowym opisie danych wyczytałem, ze body to Body Identification Number, co oznacza, ze ta kolumna powinna tyczyć się jedynie osób, które nie przezyly,
co obrazują powysze wyniki: 
- 1.000 - 100% brak danych (100% nan-ów) dla pasazerow, ktorzy przezyli (brak koniecznosci identyfikacji ciala)
- 0.850433 - 85% brak danych dla ludzi, ktorzy nie przezyli (ciała większości osób nie zostały znalezione)
>Powyzsze wyniki pozwalają wysunąć wnioski, ze mamy do czynienia z MNAR - braki w danych nie są przypadkowe.

Powyzsze dwie linie kodu zapisane w jednej:

```python
df['body'].isna().groupby(df['survived']).mean()
```

Obliczając procentowy brak danych dla całego zbioru danych, wyszedł on na poziomie 21%, co uniemozliwia nam usunięcie wszystkich wierszy z tego typu brakami.
Jak napisałem powyzej, w zbiorze mamy do czynienia z nieprzypadkowymi brakami danych, co równierz nalezy uwzględnić w analizie. W celu  usunięcia braków, nalzałoby się skupić na poszczególnych kolumnach i dla kazdej indywidualnie decydować czy mozemy usunać w niej wiersze z NaN

<h5>Komentarz wyników otrzymanych w punkcie 1 i 2 oraz podział zmiennych ze
względu na dużą i małą moc zbioru (kardynalność).</h5>

Na podstawie podpunktu 1 i 2 zauwazyłem, ze dane dziela sie na etykiety do wartości 30 oraz powyzej 99, dlatego tez według takiego kryterium podziele je na zbiory

<h5>Redukcja cech dla zmiennej opisującej kabiny poprzez zastąpienie
obecnych etykiet w formacie LL11 do etykiet zawierających tylko
pierwszą literę.</h5>

```python
df['CabinReduced'] = np.where(df['cabin'].isnull(), np.nan, df['cabin'].astype(str).str[0])
df[['cabin', 'CabinReduced']].head(20)
```
