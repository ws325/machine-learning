```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

<h5>Wczytanie przetworzonego zbioru danych utoworzoengo w pliku z lab2</h5>

```python
df = pd.read_csv('Titanic_modified.csv')
df.head(10)
```

Funkcja train_test_split słuzy do podziału danych wejściowych na dwa podzbiory - zbiór treningowy oraz testowy.<br><br>
<b>sklearn.model_selection. train_test_split ( * arrays , test_size = None , train_size = None , random_state = None , shuffle = True , stratify = None )</b>
- \* arrays: parametrem moze być lista, numpay array, scipy-sparse matrices lub pandas dataframe
- test_size: przyjmuje int - liczba próbek testowych lub float - od 0,0 do 1,0 co odpowiadać proporcji zestawu danych do podziału (na testowy i treningowy),
jeśli None to domyślnie 0.25
- train_size: przyjmuje int - liczba próbek treningowych lub float - od 0,0 do 1,0 co odpowiadać proporcji zestawu danych do podziału (na testowy i treningowy),
jeśli None to domyślnie ustawiona na dopełnienie zestawu danych testowych
- random_state: przyjmuje int - odpowiada za potasowanie danych przed podziałem na zbiory, domyślnie None


```python
col_name = ['cabin', 'CabinReduced', 'sex']
```

<h5>Podział zbioru na trengowy i testowy uzywając train_test_split.</h5>

```python
x_train, x_test, y_train, y_test = train_test_split(df[col_name], df[['survived']], test_size=0.3, random_state=0 )
```

```python
print('Wymiary: \nx_train.shape = ' + str(x_train.shape),'\nx_test.shape = ' + str(x_test.shape),
'\ny_train.shape = ' + str(y_train.shape), '\ny_test.shape = ' + str(y_test.shape))
```

>Zbiór danych testowych stanowi 30% zbioru danych wejściowych, zatem dane zostały poprawnie podzielone

<h5>Kodowanie zmiennych kategorycznych do zmiennych liczbowych.</h5>

```python
cabin_dict = {}
for num, lab in enumerate(df['cabin'].unique()):
    cabin_dict[lab] = num
cabin_dict
```

```python
CabinReduced_dict = {}
for num, lab in enumerate(df['CabinReduced'].unique()):
    CabinReduced_dict[lab] = num
CabinReduced_dict
```

```python
sex_dict = {}
for num, lab in enumerate(df['sex'].unique()):
    sex_dict[lab] = num
sex_dict
```

<h5>Uzycie mapowania do zastąpienia etykiety stworzonym słownikiem.</h5>

```python
x_test['cabin_map'] = x_test['cabin'].map(cabin_dict)
x_test
```

```python
x_test['CabinReduced_map'] = x_test['CabinReduced'].map(CabinReduced_dict)
x_test
```

```python
x_test['sex_map'] = x_test['sex'].map(sex_dict)
x_test
```

```python
x_train['cabin_map'] = x_train['cabin'].map(cabin_dict)
x_train
```

```python
x_train['CabinReduced_map'] = x_train['CabinReduced'].map(CabinReduced_dict)
x_train
```

```python
x_train['sex_map'] = x_train['sex'].map(sex_dict)
x_train
```

<h5>Sprawdzenie liczby brakujących wartości w zmodyfikowanych zbiorach.</h5>

```python
nan_sum_train = x_train.isnull().sum()
nan_sum_train
```

```python
nan_sum_test = x_test.isnull().sum()
nan_sum_test
```

Dla zmiennych, które zostały zmapowane nie ma braków danych, poniwaz Nan-om została przypisana jakaś konkretna(w zaleznosci od kolumny) wartość liczbowa
